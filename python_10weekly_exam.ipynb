```python
#사각형 클래스
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# 사용 예
rect = Rectangle(4, 5)
print("사각형의 넓이:", rect.area())

#계산기 클래스
class Calculator:
    def __init__(self):
        self.result = 0  # 결과를 저장하는 변수

    def add(self, number):
        self.result += number
        return self.result

# 계산기 두 개 생성
cal1 = Calculator()
cal2 = Calculator()

print(cal1.add(3))  # 3
print(cal1.add(4))  # 7
print(cal2.add(10)) # 10

#클래스 & 객체변수
class Counter:
    count = 0  # 클래스 변수 (공유됨)

    def __init__(self):
        self.my_count = 0  # 객체 변수 (개별적으로 다름)
        Counter.count += 1

# 객체 생성
a = Counter()
b = Counter()

print("클래스 변수:", Counter.count)  # 2
print("a 객체의 객체 변수:", a.my_count)  # 0
print("b 객체의 객체 변수:", b.my_count)  # 0

#init 함수
class Person:
    def __init__(self, name):
        self.name = name

    def say_hello(self):
        print(f"안녕하세요, 저는 {self.name}입니다.")

# 객체 생성
p1 = Person("승호")
p1.say_hello()

#사칙연산 클래스
class FourCal:
    def setdata(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second

    def subtract(self):
        return self.first - self.second

    def multiply(self):
        return self.first * self.second

    def divide(self):
        return self.first / self.second

# 사용 예
calc = FourCal()
calc.setdata(10, 5)
print("덧셈:", calc.add())
print("뺄셈:", calc.subtract())
print("곱셈:", calc.multiply())
print("나눗셈:", calc.divide())

#주사위 게임 OOP 형식
import random

class Player:
    def __init__(self, name):
        self.name = name

    def roll_dice(self):
        return random.randint(1, 6) + random.randint(1, 6) + random.randint(1, 6)

# 주사위 게임 진행
player1 = Player("A")
player2 = Player("B")

score1 = player1.roll_dice()
score2 = player2.roll_dice()

print(f"{player1.name} 점수: {score1}")
print(f"{player2.name} 점수: {score2}")

if score1 > score2:
    print(f"{player1.name} 승리!")
elif score1 < score2:
    print(f"{player2.name} 승리!")
else:
    print("무승부입니다!")
```
